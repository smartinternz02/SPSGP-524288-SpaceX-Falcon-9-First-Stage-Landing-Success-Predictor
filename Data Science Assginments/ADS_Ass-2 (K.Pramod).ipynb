# 6. Find the outliers and replace the outliers

```python
sns.boxplot(data.age)
```

```python
sns.boxplot(data.fare)
```

```python

import numpy as np
import pandas as pd

# Identify numerical columns for outlier detection
numerical_columns = ['age', 'fare']

# Define the boundaries for outliers using Tukey's fences
def find_outliers(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    outliers = (data < lower_fence) | (data > upper_fence)
    return outliers

# Replace outliers with the lower or upper fence
def replace_outliers(data):
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr
    data = np.where(data < lower_fence, lower_fence, data)
    data = np.where(data > upper_fence, upper_fence, data)
    return data

# Find and replace outliers in numerical columns
for column in numerical_columns:
    outliers = find_outliers(data[column])
    data[column] = replace_outliers(data[column])

# Updated dataframe with replaced outliers
print(data.head())
```
